```python
def describe_col(spark_df, col, bins = 50):
    """
    Affiche la distribution d'une col numerique + nb valeurs manquantes
    """
    data = spark_df.select(col).toPandas()
    
    plt.figure(figsize=(8, 5))
    plt.hist(data[col], bins=bins, edgecolor='black')
    plt.title(f"Distribution de '{col}'")
    plt.xlabel(col)
    plt.ylabel("Fr√©quence")
    plt.grid(True, alpha=0.3)
    plt.show()

    tot = spark_df.count()
    missing = spark_df.filter((spark_df[col].isNull()) | (isnan(spark_df[col]))).count()
    missing_pct = (missing / tot) * 100
    print(f"Valeurs manquantes : {missing} / {tot} ({missing_pct:.2f}%)")
```

### 1.1.1 Pr√©traitement du texte sur un √©chantillon r√©duit

CETTE PARTIE N'EST PAS UTILISEE DANS LA SUITE DU PROGRAMME JE LA LAISSE COMME ELLE PERMET DE BIEN VISUALISER LES DIFFERENTES ETAPES --> SERA SUPPRIMEE LORS DU RENDU

--> sur 5000 lignes en utilisant la librairie pandas

--> m√©thode pas adapt√©e pour le bigdata car on ram√®ne tout sur le driver pour utiliser panda

--> mais utile pour visualiser les diff√©rentes √©tapes du traitement des donn√©es

```python
from pyspark.sql.functions import lower, regexp_replace, monotonically_increasing_id

train_std = (
    train.filter(train.comment_text.isNotNull())
    .withColumn("comment_text", lower(train.comment_text))
    .withColumn("comment_text", regexp_replace("comment_text", r"[^a-zA-Z\s]", ""))
    .withColumn("comment_text", regexp_replace("comment_text", r"\s+", " "))
)
```

```python
from pyspark.sql.functions import monotonically_increasing_id

def compare_comments(df1, col1, df2, col2, n=5, truncate_len=60):
    """
    Affiche un Avant/Apr√®s pr√©traitements

    Param√®tres :
    ------------
    df1, df2 : DataFrame Spark
    col1, col2 : colonnes de DataFrame Spark
    n : int
        Nombre de lignes √† afficher (par d√©faut 5)
    truncate_len : int
        Nombre de caract√®res √† afficher avant troncature (par d√©faut 60)
    """
    
    # pour r√©duire le temps de compil
    train_limited = df1.select(col1).limit(n)
    train_std_limited = df2.select(col2).limit(n)

    # id unique sur chaque df pour jointure
    train_with_id = train_limited.withColumn("id", monotonically_increasing_id())
    train_std_with_id = train_std_limited.withColumn("id", monotonically_increasing_id())

    # inner join sur l'id
    compare_df = train_with_id.join(train_std_with_id, on="id", how="inner") \
        .select(train_with_id[col1].alias("avant"),
            train_std_with_id[col2].alias("apr√®s"))

    compare_df.show(truncate=60)

    # pour r√©utilisation √©ventuelle
    return compare_df
```

```python
compare_comments(train, "comment_text", train_std, "comment_text") # AVANT/APRES
```

**Suppression des stop words**

Stop words are the words in a stop list (or stoplist or negative dictionary) which are filtered out ("stopped") before or after processing of natural language data (i.e. text) because they are deemed to have little semantic value or are otherwise insignificant for the task at hand. 

¬´¬†Stop Word¬†¬ª. Wikipedia, 12 ao√ªt 2025. Wikipedia, https://en.wikipedia.org/w/index.php?title=Stop_word&oldid=1305565804.




```python
from pyspark.ml.feature import StopWordsRemover, Tokenizer

tokenizer = Tokenizer(inputCol="comment_text", outputCol="words")
remover = StopWordsRemover(inputCol="words", outputCol="filtered")

words_data = tokenizer.transform(train_std)
clean_data = remover.transform(words_data)
```

```python
clean_data.select("words", "filtered").show(5, truncate=60)
```

**Lemmisation**

PySpark ML n‚Äôa pas de lemmatiseur int√©gr√©, on passe donc temporairement en Pandas

```python
sample_pd = clean_data.select("filtered").limit(5000).toPandas()

import nltk
from nltk.stem import WordNetLemmatizer

nltk.download("wordnet")
nltk.download("omw-1.4")

lemm = WordNetLemmatizer()

def lemmatize_tokens(tokens):
    return [lemm.lemmatize(word) for word in tokens]

# on applique sur la colonne filtered
sample_pd["lemmatized"] = sample_pd["filtered"].apply(lemmatize_tokens)

```

```python
sample_pd["clean_text"] = sample_pd["lemmatized"].apply(lambda x: " ".join(x))

sample_spark = spark.createDataFrame(sample_pd[:10]) #on repasse en pyspark pour la comparaison

compare_comments(clean_data.limit(10), "filtered", sample_spark.limit(10), "lemmatized", n=5 ) # AVANT/APRES
```

**Variante du Lemming ‚Äî Stemming**

On tronque simplement la fin du mot --> plus rapide mais plus susceptible de perdre de l'information

```python
from nltk.stem import PorterStemmer

stemmer = PorterStemmer()

def stem_tokens(tokens):
    return [stemmer.stem(word) for word in tokens]

sample_pd["stemmed"] = sample_pd["filtered"].apply(stem_tokens)
```

```python
sample_pd.head()
```

```python
initial_count = len(sample_pd)

sample_pd = sample_pd[
    sample_pd["clean_text"].notna() &                    # non null
    sample_pd["clean_text"].astype(str).str.strip().ne("") &  # pas juste des espaces
    (sample_pd["clean_text"].astype(str) != "[]")        # pas liste vide
]

final_count = len(sample_pd)

removed = initial_count - final_count
print(f"{removed} lignes vides supprim√©es (sur {initial_count} initiales).")
```

### 1.1.2 Pr√©traitement du texte sur "l'ensemble" des commentaires

--> utilisation de fonctions UDF afin que le traitement reste distribu√© sur les workers (pas d'utilisation de pandas sauf pour l'affichage d'un √©chantillon des donn√©es !!!

--> on fait tout le pr√©traitement dans spark

--> Encapsulation dans un pipeline --> structure robuste et modulaire

```python
from pyspark.sql.functions import udf, lower, regexp_replace, monotonically_increasing_id
from pyspark.sql.types import ArrayType, StringType
from pyspark.ml import Pipeline
from pyspark.ml.feature import Tokenizer, StopWordsRemover, SQLTransformer
from nltk.stem import WordNetLemmatizer
import nltk

nltk.download("wordnet") # telechargement des ressources NLTK
nltk.download("omw-1.4")

# Ma vm est trop lente pour tout traiter ptdr. JB a toi de jouer ici !!!
# Ajout d'une col ID pour tout regrouper apr√®s
train = train.limit(10_000).withColumn("id", monotonically_increasing_id())

# Nettoyage du texte avec un SQLTransformer pour le pipeline
regex_cleaner = SQLTransformer(
    statement="""
    SELECT *,
           lower(comment_text) AS comment_text_lower,
           regexp_replace(lower(comment_text), '[^a-z ]', '') AS clean_text
    FROM __THIS__
    """
)
```

```python
tokenizer = Tokenizer(inputCol="clean_text", outputCol="words") # tokenisation = d√©couper le texte en mots (tokens)

remover = StopWordsRemover(inputCol="words", outputCol="filtered") # supprime les mots tr√®s fr√©quents et peu informatifs

# Lemmatisation --> ramene chaque mot √† sa forme canonique running --> run
lemm = WordNetLemmatizer() # vient de NLTK, c'est la base de notre UDF

def lemmatize_tokens(tokens):
    return [lemm.lemmatize(word) for word in tokens] # fonction pour lemmatiser tous les tokens d'un array

lemm_udf = udf(lemmatize_tokens, ArrayType(StringType())) #on cr√©e l'UDF, renvoie une liste de cha√Ænes de caract√®res

# On encapsule la lemmatisation dans un SQLTransformer + UDF avec withColumn
# Pour int√©grer dans le pipeline)
from pyspark.ml.feature import SQLTransformer

lemmatizer_stage = SQLTransformer(
    statement="SELECT *, filtered AS tokens_for_lemm FROM __THIS__"
)
```

```python
pipeline = Pipeline(stages=[
    regex_cleaner,
    tokenizer,
    remover
])

# ENLEVER LES COMMENTAIRES POUR VOIR LES DIFFERENTES ETAPES DU PREPROCESSING

# regex_df = regex_cleaner.transform(train)
# regex_df.select("clean_text").show(5, truncate=50)

# token_df = tokenizer.transform(regex_df)
# token_df.select("clean_text", "words").show(5, truncate=50)

# rem_df = remover.transform(token_df)
# rem_df.select("words", "filtered").show(5, truncate=50)
```

```python
preprocess_model = pipeline.fit(train) # pr√©paration du pipeline par rapport aux donn√©es != fit pour un mod√®le
preprocessed_df = preprocess_model.transform(train) #applique le pipeline √† train

preprocessed_df = preprocessed_df.withColumn("lemmatized", lemm_udf("filtered")) # on applique la lemmatisation
train_final_df = train.join(preprocessed_df.select("id", "lemmatized"), on="id", how="inner").cache() # jointure sur l'id avec le df train
```

```python
# Affichage du df reconsitu√© 
train_final_df.select("id","lemmatized", "target").limit(10).show(truncate=False)
```

### 1.1.3 Vectorisation (conversion des mots en vecteurs num√©riques)

- CountVectorizer
- TF-IDF
- Word2Vec

***CountVectorizer + IDF***

Spark ML reglog = classif binaire --> on doit convertir la colonne target en binaire pour le mod√®le baseline, meme si cela nous fait perdre de l'information

```python
from pyspark.sql.functions import when, col

train_final_df_bin = train_final_df.withColumn(
    "label",
    when(col("target") >= 0.5, 1).otherwise(0)
)

train_final_df_bin = train_final_df_bin.withColumnRenamed("lemmatized", "tokens") # pour respecter la convention sparknlp
```

```python
from pyspark.ml.feature import CountVectorizer, IDF

# listes --> sparce vector
cv = CountVectorizer(
    inputCol="tokens",
    outputCol="raw_features",
    vocabSize=100000,  # largeur voc
    minDF=5            # garde les mots qui apparaissent dans ‚â•5 docs
)

idf = IDF(
    inputCol="raw_features",
    outputCol="features"
)
```

```python
from pyspark.ml.classification import LogisticRegression

lr = LogisticRegression(
    featuresCol="features",
    labelCol="label",
    maxIter=50,
    regParam=0.0
)
```

```python
train_df, val_df, test_df = train_final_df_bin.randomSplit([0.7, 0.15, 0.15], seed=50)
```

```python
pipeline = Pipeline(stages=[cv, idf, lr])

model = pipeline.fit(train_df)

preds = model.transform(test_df)
preds.select("tokens","target","probability","prediction").show(truncate=60)
```

```python
from pyspark.ml.evaluation import BinaryClassificationEvaluator

evaluator = BinaryClassificationEvaluator(
    labelCol="target",
    rawPredictionCol="rawPrediction",
    metricName="areaUnderROC"
)

auc = evaluator.evaluate(preds)
print("Baseline ROC-AUC =", auc)
```
